```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_data = np.random.random(1000)*100
y_data = np.random.random(1000)*100
```

### Scatter

```python
plt.scatter(X_data, y_data, color='#225490', marker="*", s=120, alpha=0.5)
# alpha = opacity


plt.show()
```

### Bar

```python
x = ["C", "Python", "Java", "JavaScript", "C++", "Rust"]
y = [15, 21, 27, 25, 38, 12]
```

```python
plt.bar(x,y, color="green", align="center", width=0.6, edgecolor="red", linewidth=3) # align="edge"

plt.show()
```

### Histograms

```python
ages = np.random.normal(20, 3, 100)
ages
```

```python
plt.hist(ages, bins=30) # acumulative=True

plt.show()
```

### Pie Chart

```python
langs = ["C", "Python", "Java", "JavaScript", "C++", "Rust"]
populations = [15, 21, 27, 25, 38, 12]
explodes = [0, 0, 0, 0, 0.1, 0] # drag size 
```

```python
plt.pie(populations, labels=langs, explode=explodes, autopct="%.2f%%", shadow=False, startangle=0) #pctdistance=0.85

plt.show()
```

### Customization

```python
years = [1996 + i for i in range(9)]
income = [80, 81, 82, 94, 95, 85, 86, 87, 88]
income_ticks = list(range(80, 100, 2))
```

```python
plt.plot(years, income)

plt.title("Income of company (USD)", fontsize=21, fontname="impact")
plt.xlabel("Years")
plt.ylabel("Income (USD)")
plt.yticks(income_ticks, [f"${i}k" for i in income_ticks])

plt.show()
```

```python
stock_A = [100, 81, 93, 99, 80, 89, 91, 92, 88] #only x values, y generated automacally
stock_B = [88, 90, 100, 94, 95, 96, 87, 86, 93]
stock_C = [90, 99, 80, 85, 100, 99, 81, 94, 87]
```

```python
plt.plot(stock_A, label="Stock A")
plt.plot(stock_B, label="Stock B")
plt.plot(stock_C, label="Stock C")

plt.legend(loc="lower right")

plt.show()
```

### Multiple figures

```python
X1, y1 = np.random.random(100), np.random.random(100)
X2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(X1, y1)

plt.figure(2)
plt.plot(X2, y2)

plt.show()
```

### Subplots

```python
fig, axs = plt.subplots(2,2)

x = np.arange(100)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("sin(x)")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("cos(x)")

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("Random(x)")

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("log(x)")

fig.suptitle("Four plots")

plt.tight_layout()

# plt.show()

# Save pltos
plt.savefig("fourplots.png", dpi=300, transparent=False)
```
